```python
from crewai import Agent

Market_Analyst = Agent(
    role = 'Market Research Analyst',
    goal = 'Research market trends, size, and target audience',
    backstory= 'You are a market analyst with 5 years of experience in the tech industry. You have a strong background in data analysis and market research.',
    max_iter=2,
    verbose=True
)

Competitor_Scout = Agent(
    role = 'Competitive Intelligence Analyst',
    goal = 'Identify and summarize key competitors',
    backstory= 'You are a competitor scout with 3 years of experience in the tech industry. You have a strong background in competitive analysis and market research.',
    max_iter=2,
    verbose=True
)

Feasibility_Analyst = Agent(
    role = 'Feasibility Analyst',
    goal = 'Assess technical/business feasibility of the idea',
    backstory= 'You are a feasibility analyst with 4 years of experience in the tech industry. You have a strong background in product development and market research.',
    max_iter=2,
    verbose=True
)

Business_Writer = Agent(
    role = 'Business Writer',
    goal = '	Draft a one-pager executive summary',
    backstory= 'You are a business writer with 6 years of experience in the tech industry. You have a strong background in business writing and market research.',
    max_iter=2,
    verbose=True

)

Pitch_Designer = Agent(
    role = 'Pitch Designer',
    goal = 'Create a pitch deck outline with slide titles and content summaries',
    backstory= 'You are a pitch designer with 5 years of experience in the tech industry. You have a strong background in design and market research.',
    max_iter=2,
    verbose=True
)
```

```python
from crewai import Task
startup_idea = "AI-powered personal finance assistant for Gen Z"


research_task = Task(
    description=f"Conduct in-depth research on market trends, size, and target audience for the startup idea: {startup_idea}.",
    expected_output="Comprehensive market research report.",
    agent=Market_Analyst,
)


competitor_task = Task(
    description=f"Identify key competitors in the space and analyze their offerings for the startup idea: {startup_idea}.",
    expected_output="Detailed competitor analysis report.",
    agent=Competitor_Scout,
)


feasibility_task = Task(
    description="Evaluate the technical and business feasibility of the startup idea using market and competitor analysis.",
    expected_output="Feasibility report evaluating potential challenges and opportunities.",
    agent=Feasibility_Analyst,
    context=[research_task, competitor_task]
)


business_task = Task(
    description="Write a one-pager business summary using the feasibility report.",
    expected_output="One-pager executive summary for the startup idea.",
    agent=Business_Writer,
    context=[feasibility_task]
)


pitch_task = Task(
    description="Create a pitch deck outline with slide titles and content summaries based on the business summary.",
    expected_output="Complete pitch deck outline with 10-12 slides.",
    agent=Pitch_Designer,
    context=[business_task]
)
```

```python
from crewai import Crew

crew = Crew(
    agents=[Market_Analyst,Competitor_Scout,Feasibility_Analyst,Business_Writer,Pitch_Designer],
    tasks=[research_task,competitor_task,feasibility_task,business_task,pitch_task],
    erbose=True
)


result = crew.kickoff()
print(result)
```
